```python
#Используя библиотеку NumPy, создать матрицу 5x5 случайных вещественных чисел, принадлежащих интервалу (0, 2). Транспонировать. Вычислить ее определитель.
matrix_a = 2 * np.random.rand(5, 5)
print("Исходная матрица A:\n", matrix_a)

matrix_a_transposed = matrix_a.T
print("\nТранспонированная матрица A^T:\n", matrix_a_transposed)

det_a = np.linalg.det(matrix_a_transposed)
print(f"\nОпределитель матрицы A^T: {det_a:.4f}")
```

```python
# Используя библиотеку NumPy, создать вектор-столбец и матрицу подходящих размеров. Выполнить умножение матриц.

# Создаем матрицу 3x2
matrix_b = np.array([[1, 2], [3, 4], [5, 6]])
print("Матрица B (3x2):\n", matrix_b)

# Создаем вектор-столбец 2x1
vector_v = np.array([[7], [8]])
print("\nВектор-столбец v (2x1):\n", vector_v)

# Умножаем матрицу на вектор. Результат - матрица 3x1.
result = np.dot(matrix_b, vector_v) # или matrix_b @ vector_v
print("\nРезультат умножения B * v (3x1):\n", result)
```

```python
# Используя библиотеку SymPy, упростите выражение bнайдите его значение

import sympy as sp
import math

x, y = sp.symbols('x y')

expr = (2/x**2 + 3/y**2) * (x + 3*y)
print("Исходное выражение:", expr)

simplified_expr = sp.simplify(expr)
print("Упрощенное выражение:", simplified_expr)

x_val = -2.01
y_val = math.sqrt(5)

result = simplified_expr.subs({x: x_val, y: y_val})
print(f"\nЗначение выражения при x={x_val}, y=√5: {result:.5f}")
```

```python
# Используя библиотеку SymPy, найдите частные производные от выражения из задания выше.
x, y = sp.symbols('x y')

# Находим частную производную по x
derivative_x = sp.diff(simplified_expr, x)
print("\nЧастная производная по x:")
sp.pprint(derivative_x) # Красивый вывод

# Находим частную производную по y
derivative_y = sp.diff(simplified_expr, y)
print("\nЧастная производная по y:")
sp.pprint(derivative_y)
```

```python
# Вычислить интеграл двумя способами: используя библиотеку SciPy и библиотеку SymPy
import scipy.integrate as spi
# SciPy численно
# Определяем подынтегральную функцию
def integrand_func(x):
    return 1 / (1 + np.sqrt(2*x + 1))

# Вычисляем интеграл от 0 до 4
result_scipy, error_estimate = spi.quad(integrand_func, 0, 4)
print(f"Результат SciPy (quad): {result_scipy:.6f}")

#SymPy символьно
x = sp.symbols('x')
expr = 1 / (1 + sp.sqrt(2*x + 1))
# Вычисляем неопределенный интеграл (первообразную)
indef_integral = sp.integrate(expr, x)
print(f"\nПервообразная (SymPy): {indef_integral}")

#Вычисляем определенный интеграл
def_integral = sp.integrate(expr, (x, 0, 4))
print(f"Точное значение (SymPy): {def_integral}")

#Получаем численное значение
result_sympy = def_integral.evalf()
print(f"Численное значение (SymPy): {result_sympy:.6f}")
```

```python
import sympy as sp
import scipy.integrate as spi
import numpy as np


#SymPy символьный
x, y = sp.symbols('x y')
expr = sp.cos(y + x)

#Внутренний интеграл по y
inner = sp.integrate(expr, (y, 0, x + sp.pi))
print(f"Внутренний интеграл по y: {inner}")

# Внешний интеграл по x
result_sym = sp.integrate(inner, (x, 0, sp.pi/2))
print(f"\nТочное значение (SymPy): {result_sym}")
print(f"Численное значение: {result_sym.evalf():.6f}")

#SciPy численный
def integrand(y, x):
    """Подынтегральная функция cos(y+x)"""
    return np.cos(y + x)

result_num, error = spi.dblquad(
    integrand,
    0, np.pi/2,                    # пределы по x
    lambda x: 0,                   # нижний предел по y
    lambda x: x + np.pi            # верхний предел по y
)

print(f"\nЧисленное значение (SciPy): {result_num:.6f}")
```

```python
# Построить в одной системе координат графики функций: y=ln(x+5), y=3x-2 Оси координат должны быть подписаны, графики должны быть разного цвета, должна быть выведена легенда. Точку пересечения (если она есть) выделить на графике оранжевым цветом.

import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def f1(x):
    return np.log(x + 5)

def f2(x):
    return 3*x - 2

# Функция для поиска пересечения (f1(x) - f2(x) = 0)
def intersection_func(x):
    return f1(x) - f2(x)


x_intersect = fsolve(intersection_func, 0)[0]
y_intersect = f1(x_intersect)

print(f"Точка пересечения: x = {x_intersect:.4f}, y = {y_intersect:.4f}")

# Создаем массив x. Область определения логарифма: x > -5
x_vals = np.linspace(-4.5, 3, 400)
y1_vals = f1(x_vals)
y2_vals = f2(x_vals)

# Построение графиков
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y1_vals, 'b-', linewidth=2, label='y = ln(x+5)')
plt.plot(x_vals, y2_vals, 'g--', linewidth=2, label='y = 3x - 2')

# Выделяем точку пересечения
plt.scatter(x_intersect, y_intersect, color='orange', s=100, zorder=5, label=f'Пересечение ({x_intersect:.3f}, {y_intersect:.3f})')

# Оформление графика
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Графики функций y = ln(x+5) и y = 3x - 2')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-4.5, 3)
plt.ylim(-5, 10)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Настройка фигуры
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Параметры для формы шишки
# u - угол вокруг оси (0 до 2*pi)
# v - высота (от 0 до 1)
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, 1, 50)
U, V = np.meshgrid(u, v)

# Математика формы шишки
# 1. Базовый радиус уменьшается к верху (форма конуса): (1 - V)
# 2. Добавляем спираль: угол зависит от высоты (V * 10)
# 3. Добавляем чешуйки: синусоида по углу и высоте
radius = (1 - V) * (1 + 0.4 * np.sin(15 * U + 20 * V))

# Преобразование в декартовы координаты (X, Y, Z)
# Поворачиваем на 90 градусов, чтобы шишка стояла вертикально
X = radius * np.cos(U)
Y = radius * np.sin(U)
Z = V * 2.5  # Умножаем на 2.5, чтобы вытянуть по высоте

# Рисуем поверхность
# cmap='brown' дает коричневый цвет, alpha=0.9 делает её чуть прозрачной для красоты
surf = ax.plot_surface(X, Y, Z, cmap='terrain', linewidth=0, antialiased=True, alpha=0.95)

# Добавим немного "хвои" или стебля снизу (простой цилиндр)
z_stem = np.linspace(-0.5, 0, 20)
theta_stem = np.linspace(0, 2 * np.pi, 20)
Z_stem, Theta_stem = np.meshgrid(z_stem, theta_stem)
X_stem = 0.15 * np.cos(Theta_stem)
Y_stem = 0.15 * np.sin(Theta_stem)
ax.plot_surface(X_stem, Y_stem, Z_stem, color='#5D4037', alpha=1)

# Настройка отображения
ax.set_title('3D Pine Cone (Шишка)', fontsize=15, fontweight='bold')
ax.set_axis_off() # Скрываем оси для красоты

# Настройка камеры для лучшего вида
ax.view_init(elev=20, azim=-60)

# Отображение графика
plt.show()
```
